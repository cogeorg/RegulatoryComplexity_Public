```python
import os
import pandas as pd
import re

# Step 1: Define file paths
in_file_name = os.path.expanduser('~/Git/RegulatoryComplexity_Public/310_EBA-clean/Annex5_textonly.txt')
ngram_file_name = os.path.expanduser('~/Git/RegulatoryComplexity_Public/200_analysis/ngrams_remove.csv')
out_file_name = os.path.expanduser('~/Git/RegulatoryComplexity_Public/310_EBA-ngrams/Annex5_textonly-6.txt')

# Step 2: Read the contents of the input file
with open(in_file_name, 'r') as file:
    in_text = file.read()

# Step 3: Read the strings to be removed from bla.csv (assuming a single column)
df_out = pd.read_csv(ngram_file_name, header=None)
strings_to_remove = df_out[0].tolist()

print("Before removing: ", len(in_text))

# Step 4: Initialize a dictionary to store the count of occurrences for each string
occurrences = {}

# Step 5: Loop through each string in bla.csv and remove it from in_text (case insensitive)
for string_ in strings_to_remove:
    # Count occurrences of the string (case insensitive)
    count = len(re.findall(re.escape(string_), in_text, flags=re.IGNORECASE))
    
    # Save the count of occurrences
    occurrences[string_] = count
    
    # Remove all occurrences (case insensitive)
    in_text = re.sub(re.escape(string_), '', in_text, flags=re.IGNORECASE)

# Step 6: Print the number of occurrences for each string
for string_, count in occurrences.items():
    print(f"n-gram '{string_}' found {count} times and removed.")

# Step 7: Save the modified text
with open(out_file_name, 'w') as file:
    file.write(in_text)

print("After removing: ", len(in_text))
print(f"Text processing complete. Cleaned text saved to {out_file_name}")
```

```python
# Import necessary libraries
import pandas as pd
import re

# Step 1: Read the Excel file foo.xlsx
foo_df = pd.read_excel('/Users/pierregeorg/Git/RegulatoryComplexity_Public/200_analysis/ngrams_output-3-5.xlsx')

# Step 2: Clean up the first column by removing characters like ( ) ' , and convert to lowercase
def clean_string(value):
    # Remove special characters (, ), ', and ,
    cleaned_value = re.sub(r"[()'\",]", "", value)
    # Convert to lowercase
    return cleaned_value.lower()

# Apply the clean_string function to the first column in foo.xlsx
foo_df['cleaned_column'] = foo_df.iloc[:, 0].apply(clean_string)

# Step 3: Read the CSV file bla.csv
bla_df = pd.read_csv('/Users/pierregeorg/Git/RegulatoryComplexity_Public/200_analysis/results/Master_v2.0-BE.csv', sep=';')

# Step 4: Store the values of the first column of bla.csv into a list
key_list = bla_df['key'].str.lower().tolist()

# Step 5: Check if the cleaned first column values from foo.xlsx exist in the key list
matches = foo_df[foo_df['cleaned_column'].isin(key_list)]

# Step 6: Drop matching rows from foo_df
foo_df_filtered = foo_df[~foo_df['cleaned_column'].isin(key_list)]

# Step 7: Print the number of rows in foo.xlsx and the number of matches found
original_row_count = len(foo_df)
match_count = len(matches)
remaining_row_count = len(foo_df_filtered)

print(f"Number of rows in foo.xlsx: {original_row_count}")
print(f"Number of matches found: {match_count}")
print(matches)
print(f"Number of rows after dropping matches: {remaining_row_count}")

# Optional: Save the filtered data to a new Excel file
foo_df_filtered.drop(columns=['cleaned_column'], inplace=True)  # Drop the cleaned column if not needed
foo_df_filtered.to_excel('/Users/pierregeorg/Git/RegulatoryComplexity_Public/200_analysis/ngrams_output-3-5-filtered.xlsx', index=False)
```
